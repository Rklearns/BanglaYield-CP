### This dataset has no missing values, so no further pre processing for missing needed

```python
reduced_df = data.drop(
    ['Production'], axis = 1)
```

```python
reduced_df.head()
```

```python
reduced_df.Transplant.value_counts()
```

```python
month_dict = {'January':'Jan', 'February':'Feb', 'March':'Mar', 'April':'Apr', 'May':'May', 'June':'Jun', 'July':'Jul', 'August':'Aug', 'September':'Sep', 'October':'Oct', 'November':'Nov', 'December':'Dec'}
for key in reduced_df.Transplant.unique():
    if key in month_dict:
        reduced_df.Transplant.replace(key, month_dict[key], inplace=True)
```

```python
reduced_df.Transplant.value_counts()
```

```python
reduced_df.Growth.value_counts()
```

```python
for key in reduced_df.Growth.unique():
    start, end = key.split('to')
    if start.strip() in month_dict: 
        start = month_dict[start.strip()]
    if end.strip() in month_dict:
        end = month_dict[end.strip()]

    new_key = start + ' to ' + end
    reduced_df.Growth.replace(key, new_key, inplace=True)

reduced_df.Growth.value_counts()
```

```python
reduced_df.Harvest.value_counts()
```

```python
for key in reduced_df.Harvest.unique():
    if 'to' not in key:
        if key.strip() in month_dict:
            reduced_df.Harvest.replace(key, month_dict[key.strip()], inplace=True)
        continue
    start, end = key.split('to')
    if start.strip() in month_dict: 
        start = month_dict[start.strip()]
    if end.strip() in month_dict:
        end = month_dict[end.strip()]

    new_key = start + ' to ' + end
    reduced_df.Harvest.replace(key, new_key, inplace=True)

reduced_df.Harvest.value_counts()
```
